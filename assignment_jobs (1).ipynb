```python
#getting the title,company_name,location,state,category,subcategory,company_link
title=[]
company_name=[]
location=[]
state=[]
category_=[]
sub_category=[]
company_link=[]
for i in range(1): #data
  category=data[i]['category']
  subcategory=data[i]['subcategory']
  for j in range(len(subcategory)): #subcat
   for k in range(len(states_list)): 
    url = 'https://www.linkedin.com/jobs/search/?keywords='+subcategory[j].replace(' ','%20')+'&location='+states_list[k].replace(' ',"%20")#modifing the url
    driver.get(url)                                                                                    #to get the link           
    time.sleep(15)
    if driver.find_elements_by_class_name("jobs-search-two-pane__no-results-banner--expand"):
        continue
    jobs = driver.find_elements_by_css_selector('.artdeco-entity-lockup__content')
    driver.execute_script("window.scrollTo(0,document.body.scrollHeight)")
    time.sleep(3)
    print(len(jobs))
       
    for l in range(len(jobs)):
     
        title.append(jobs[l].find_element_by_class_name('job-card-list__title').text)#getting the title
        company_name.append(jobs[l].find_element_by_class_name('job-card-container__company-name').text)#getting the company_name
        company_link.append(jobs[l].find_element_by_class_name('job-card-container__company-name').get_attribute('href'))
        loc = jobs[l].find_element_by_class_name('job-card-container__metadata-item').text#getting the company_location
        location.append(loc)
        state.append(states_list[k])    
        category_.append(category)   #getting the category
        sub_category.append(subcategory[j])   #getting the subcategory      
```

```python
#creating  the pandas dataframe to store the list from above cell
df = pd.DataFrame({'title' : title,
                                'company_name' : company_name,
                                'location' : location,
                                'state': state,
                                'category_name':category_,
                                 'sub_category':sub_category,
                                'company_link': company_link
                                 }, 
                                columns=['title','company_name','location', 'state','category_name','sub_category','company_link'])

df     
```

```python
#getting the about,company_size,heeadquarters
about=[]
company_size=[]
headquarters=[]

for i in range (len(df['company_link'])):
 if df['company_link'][i]==None:
    about.append('-')
    headquarters.append('-')
    company_size.append('-')
    continue
 driver.get(df['company_link'][i]+'/about')
 time.sleep(5)
 if not driver.find_elements_by_css_selector('.text-heading-large+p'):
    about.append('-')
 else:
    about.append(driver.find_element_by_css_selector('.text-heading-large+p').text) #getting the about
    
 if not driver.find_elements_by_xpath("//dt[text()='Headquarters']/following-sibling::dd"):
    headquarters.append('-')
 else:
    headquarters.append(driver.find_element_by_xpath("//dt[text()='Headquarters']/following-sibling::dd").text) #getting the headquarters
 if not driver.find_elements_by_xpath("//dt[text()='Company size']/following-sibling::dd"):
    company_size.append('-')
 else:
    company_size.append(driver.find_element_by_xpath("//dt[text()='Company size']/following-sibling::dd").text)
    #getting the company size
    #the if else statements are necessary to check whether the company profile has all the attributes or not 
```

```python
#creating another dataframe to store the 3 list from above cell
df2=pd.DataFrame({'about' : about,
                                'headquarters' : headquarters,
                                'company_size' : company_size
                                 }, 
                                columns=['about','headquarters','company_size'])
df2
```

```python
final = pd.concat([df,df2],axis=1) #cconcating bothe the dataframe from above to get final result
```
